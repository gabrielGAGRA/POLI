## Classes abstratas e interfaces

### Classes Abstratas:

 Indica que um tipo de objeto vai possuir um determinado mtodo, mas no define a implementao do mtodo.

 Exemplo:
     Classe Filme
     Subclasses:
         FilmeMP4, FilmeAVI, FilmeOgg

### Interface:

 Funciona como uma classe puramente abstrata.  
 Lista de assinaturas de mtodos.  
 Sem implementao.  
 Deve ser implementada posteriormente por classes, usando implements.  
 class Cx implements InterfaceY  
 Todas variveis so implicitamente public static final  
 Exemplo:  
     Interface ObjetoEspacial  
         classes Asteroide, Nave, tiro, Planeta  

 Java no permite herana mltipla de classes (so 1 extends) mas permite de interfaces.

## Polimorfismo

Polimorfismo Paramtrico:

Sobrecarga de Mtodos:

 Method overloading  
 2 ou mais mtodos com o mesmo nome  
 int multiplica (int a, int b)
 float multiplica (float a, float b)  
 Complexo multiplica (Complexo a, Complexo b)

Sobreescrita de mtodo:
